# **Data Analysis**

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# read the clean data csv file
data = pd.read_csv('/content/Q2_clean_data.csv')
```

```python
data.head(15)
```

```python
#Drop the Unmamed column
data = data.drop(data.columns[0], axis=1)
data.head()
```

```python
# Group the table by 'lunch' column and count the number of student for each lunch choice
lunch_ratio = data.groupby(['lunch'])['lunch'].count()
lunch_ratio
```

```python
#Draw a pie chart to show the ratio of standard and free/reduced lunch
plt.figure(figsize=(8,10))
mylabels = ['free/reduced', 'standard']
plt.pie(lunch_ratio, labels = mylabels, autopct='%1.2f%%')
plt.legend()
plt.title('The ratio of student lunch plan', fontsize=20)
plt.show() 
```

Thanks to this pie chart, we can clearly see there is 35.5% of students having free/reduced lunch and 64.5% student having standard meal

```python
# Group the table by 'parental_level_of_education' column and count the number of parent for each level of education
education_ratio = data.groupby(['parental_level_of_education'])['parental_level_of_education'].count()
education_ratio
```

```python
# Draw the bar graph to show the parental level of education
fig =plt.figure(figsize=(12,12))
values = ['associoate\'s degree', 'bachelor\'s degree', 'high school', 'master\'s degree', 'some college', 'some high school']
plt.bar(values, education_ratio, color=['red','yellow','green','blue','orange','purple'])
plt.title('Parental Level of Education')
plt.show()
```

```python
# Group the table by 'gender' and 'test_preparation_course' columns and count the number of male and female students complete and not complete the rest preparation course
test_preparation_course_ratio = data.groupby(['gender','test_preparation_course'])['test_preparation_course'].count()
test_preparation_course_ratio
```

```python
# Draw a pie chart to show the number of female students complete test preparation course
plt.figure(figsize=(8,10))
mylabels = ['completed', 'none']
values = [test_preparation_course_ratio[0], test_preparation_course_ratio[1]]
plt.pie(values, labels = mylabels, autopct='%1.2f%%')
plt.title('The ratio of female students completing test preparation course', fontsize=20)
plt.legend()
plt.show() 
```

```python
# Draw a pie chart to show the number of male students complete test preparation course
plt.figure(figsize=(8,10))
mylabels = ['completed', 'none']
values = [test_preparation_course_ratio[2], test_preparation_course_ratio[3]]
plt.pie(values, labels = mylabels, autopct='%1.2f%%')
plt.title('The ratio of male students completing test preparation course', fontsize=20)
plt.legend()
plt.show() 
```

From the aboove charts, we clearly see out of 518 female students, there is only 35.52% students complete the test preparation course. On the other hand, there is 36.10% of male students complete the test preparation course.










```python
# create an average score column
data['average_score'] = data[['math_score', 'reading_score', 'writing_score']].mean(axis=1).round(2)
data.head(15)
     
```

```python
# Group the table by 'gender' column and count the average score of male and femal students 
average_ratio = data.groupby(['gender'])['average_score'].mean()
average_ratio
```

```python
# Draw the bar graph to show the parental level of education
fig =plt.figure(figsize=(12,12))
values = ['female', 'male']
plt.bar(values, average_ratio, color=['red','blue'])
plt.title('The average score of male and female students')
plt.show()
```

From this bar graph, we can see female students have higher average score than male students
